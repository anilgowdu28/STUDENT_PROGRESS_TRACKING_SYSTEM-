```python
print("STUDENT PROGRESS TRACKER")
sname=input("Enter the Student Name: ")
Year=int(input("Enter the Year(1/2/3/4): "))
sem=int(input("Enter the Semester(1/2): "))
Rollno=input("Enter the Rollno: ")


valid_branches = ["AI&DS", "CSC", "CSD", "CSM", "CAI"]
while True:
    Branch = input("Enter the Branch (AI&DS / CSC / CSD / CSM / CAI): ").upper()
    if Branch in valid_branches:
        break
    else:
        print("Invalid Branch! Please select from: AI&DS, CSC, CSD, CSM, CAI")

entered_marks1=int(input("Enter the Subject-1 Marks: "))
entered_marks2=int(input("Enter the Subject-2 Marks: "))
entered_marks3=int(input("Enter the Subject-3 Marks: "))
entered_marks4=int(input("Enter the Subject-4 Marks: "))
entered_marks5=int(input("Enter the Subject-5 Marks: "))

#Checking the each subject result.
def subject_result(marks):
    return "Pass" if marks >= 35 else "Fail"
sub1_result = subject_result(entered_marks1)
sub2_result = subject_result(entered_marks2)
sub3_result = subject_result(entered_marks3)
sub4_result = subject_result(entered_marks4)
sub5_result = subject_result(entered_marks5)

#Total marks.
total_marks=entered_marks1+entered_marks2+entered_marks3+entered_marks4+entered_marks5
#Average.
average=total_marks/5

#Grade Validation
if average >= 90:
    grade = "A+"
elif average >= 75:
    grade = "A"
elif average >= 65:
    grade = "B"
elif average >= 55:
    grade = "C"
elif average >= 45:
    grade = "D"
elif average >= 35:
    grade = "E"
else:
    grade = "F"

#Final result.
if(sub1_result == "Pass" and sub2_result == "Pass" and sub3_result == "Pass" and sub4_result == "Pass" and sub5_result == "Pass"):
    final_result = "PASS"
else:
    final_result = "FAIL"


print("\n..... STUDENT PROGRESS REPORT .....")
print(f"Name           : {sname}")
print(f"Branch         : {Branch}")
print(f"Year           : {Year}")
print(f"Semester       : {sem}")
print(f"Roll Number    : {Rollno}")
print("....................................")
print("Marks Obtained:")
print(f"Subject 1      : {entered_marks1}      {sub1_result}")
print(f"Subject 2      : {entered_marks2}      {sub2_result}")
print(f"Subject 3      : {entered_marks3}      {sub3_result}")
print(f"Subject 4      : {entered_marks4}      {sub4_result}")
print(f"Subject 5      : {entered_marks5}      {sub5_result}")
print("....................................")
print(f"Total Marks    : {total_marks}/500")
print(f"Average        : {average:.2f}")
print(f"Grade          : {grade}")
print(f"Final Result   : {final_result}")
print("....................................")
print("....................................")

```
